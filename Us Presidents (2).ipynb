```python
# Peforming Desciprive Statistics of the Presidents Heights

print('Mean of heights = ', height.mean())
print('Standard Deviation of heights = ', height.std())
print('Maximum heights = ', height.max())
print('Minimum heights = ', height.min())
```

```python
# Exploring Percentiles 

print('25th Percentile =', np.percentile(height, 25))
print('Median =', np.percentile(height, 50))
print('75th Percentile =', np.percentile(height, 75))
print('90th Percentile =', np.percentile(height, 90))
```

```python
## Plotting a Graph of Presidents Heights
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.hist(height, density=1, facecolor='g', alpha=0.75)
plt.title('Heights of USA Presidents')
plt.xlabel("Heights(cm)")
plt.ylabel("Number")
```

```python
# Line Graph of Heights over time of Presidency

plt.plot(order, height, color='red', marker='o')
plt.ylabel('Heights')
plt.xlabel('Order of Presidency from First to Last')
plt.show()
```
